# Limpiar los tweets
Los tweets contienen objetos innecesarios tales como: hastags, links y puntuación, la cual puede afectar el rendimiento de los algoritmos, por lo tanto eliminamos estos elementos. Todos los textos son convertidos a minusculas, para eitar que el algoritmo interprete la misma palabra como diferente.

```python
def text_preprocess(sen): 

    sen = re.sub('[^a-zA-Z]', ' ', sen)
    sen = re.sub(r"\s+[a-zA-Z]\s+", ' ', sen)
    sen = re.sub(r'\s+', ' ', sen)
    sen = sen.lower()
    return sen
```

```python
data['Embedded_text'] = data['Embedded_text'] .apply(text_preprocess)
```

```python
X = data["Embedded_text"]  
X_tweets = []  
messages = list(X)  
for mes in messages:  
    X_tweets.append(text_preprocess(mes)) 
```

# Tokenización, lematización y eliminación de palabras vacías
Las palabras vacías son palabras de uso común cuya presencia en una oración tiene menos peso en comparación con otras palabras. Incluyen palabras como 'y', 'o', 'tiene', etc.

La tokenización es el proceso de dividir una cadena en una lista de tokens. 

La lematización consiste en reducir una palabra a su forma raíz.

```python
import en_core_web_sm
nlp = en_core_web_sm.load() 
tokenizer = RegexpTokenizer(r'\w+')
lemmatizer = WordNetLemmatizer()
stop = set(stopwords.words('spanish'))
punctuation = list(string.punctuation)
stop.update(punctuation)
w_tokenizer = WhitespaceTokenizer()

# Limpiamos el set de palabras
            
def furnished(text):
    final_text = []
    for i in text.split():
        if i.lower() not in stop:
            word = lemmatizer.lemmatize(i)
            final_text.append(word.lower())
    return (final_text)
```

Debido a que los algoritmos son basados en matematicas trabajan con numeros, es necesario convertir el texto de los tweets y de las palabras claves en forma numerica. Para esto se utilizo TfidfVectorizer class from the sklearn.feature_extraction


```python
def get_vectors(*strs):
    text = [t for t in strs]
    vectorizer = TfidfVectorizer(text)
    vectorizer.fit(text)
    return vectorizer.transform(text).toarray()
```

Se utiliza Jaccard similarity, este es bueno para casos en los que duplicación no importa

```python
def jaccard_similarity(query, document):
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    return len(intersection)/len(union)
# jaccard_score(socialvector, economic_vector)

#for similarity of 1 and 2 of column1
# jaccard_similarity('dog lion a dog','dog is cat')


def get_scores(group,tweets):
    scores = []
    for tweet in tweets:
        s = jaccard_similarity(group, tweet)
        scores.append(s)
    return scores
```
